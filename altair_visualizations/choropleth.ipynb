```python
import altair as alt
import pandas as pd
from vega_datasets import data
```

```python
# orig altair choropleth
counties = alt.topo_feature(data.us_10m.url, 'counties')
unemp_data = data.unemployment.url

alt.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(unemp_data, 'id', ['rate'])
).properties(
    projection={'type': 'albersUsa'},
    width=500, height=300
)
```

```python
# read data and rename columns
lfpr_data = pd.read_csv('Labor-force-participation-map-county.csv', header = 1)
lfpr_data = lfpr_data.rename(columns={
    "County name": "county",
    "Labor force participation rate: total": "total",
    "Labor force participation rate: men": "men",
    "Labor force participation rate: women": "women"
})
lfpr_data.head()
```

```python
counties = alt.topo_feature(data.us_10m.url, 'counties')

# plot of aggregate lfpr
total = alt.Chart(counties).mark_geoshape().encode(
    color = alt.Color("total:Q", legend = alt.Legend(title = "Percent"))
).transform_lookup(
    lookup = "id",
    from_ = alt.LookupData(data = lfpr_data, key = "FIPS", fields = ["total"])
).properties(
    projection={'type': 'albersUsa'},
    width=500, height=300,
    title = "2019 Labor Force Participation Rate"
)
total
```

```python
# plot of male lfpr
men = alt.Chart(counties).mark_geoshape().encode(
    color = alt.Color("men:Q", legend = alt.Legend(title = "Percent"))
).transform_lookup(
    lookup = "id",
    from_ = alt.LookupData(data = lfpr_data, key = "FIPS", fields = ["men"])
).properties(
    projection={'type': 'albersUsa'},
    width=500, height=300,
    title = "2019 Labor Force Participation Rate: Men"
)

men
```

```python
# plot of female lfpr
women = alt.Chart(counties).mark_geoshape().encode(
    color = alt.Color("women:Q", legend = alt.Legend(title = "Percent"))
).transform_lookup(
    lookup = "id",
    from_ = alt.LookupData(data = lfpr_data, key = "FIPS", fields = ["women"])
).properties(
    projection={'type': 'albersUsa'},
    width=500, height=300,
    title = "2019 Labor Force Participation Rate: Women"
)

women
```

```python
# vertical concatenation of the three choropleths
total & men & women
```
